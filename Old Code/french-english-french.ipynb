```python
from transformers import MarianTokenizer, MarianMTModel
import speech_recognition as sr
import time


# Load translation models
fr_to_en_model_name = "Helsinki-NLP/opus-mt-fr-en"
en_to_fr_model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer_fr_to_en = MarianTokenizer.from_pretrained(fr_to_en_model_name)
model_fr_to_en = MarianMTModel.from_pretrained(fr_to_en_model_name)

tokenizer_en_to_fr = MarianTokenizer.from_pretrained(en_to_fr_model_name)
model_en_to_fr = MarianMTModel.from_pretrained(en_to_fr_model_name)

```

French-English-French

```python
while True:
    r = sr.Recognizer()

    with sr.Microphone() as source:
        print("Speak something in French...")
        audio_text = r.listen(source, timeout=15)  # Listen for 15 seconds

    try:
        text_fr = r.recognize_google(audio_text, language="fr-FR")
        print('Converting audio transcripts into French text ...')
        print(text_fr)
        spoken_french_text = text_fr
        print()
        
        # Perform translation from French to English
        input_ids_fr_to_en = tokenizer_fr_to_en.encode(spoken_french_text, return_tensors="pt")
        output_ids_en = model_fr_to_en.generate(input_ids_fr_to_en)
        translated_output_en = tokenizer_fr_to_en.decode(output_ids_en[0], skip_special_tokens=True)
        
        print("Translated to English:", translated_output_en)

        # Translate English back to French
        input_ids_en_to_fr = tokenizer_en_to_fr.encode(translated_output_en, return_tensors="pt")
        output_ids_fr = model_en_to_fr.generate(input_ids_en_to_fr)
        translated_output_fr = tokenizer_en_to_fr.decode(output_ids_fr[0], skip_special_tokens=True)
        
        print("Translated back to French:", translated_output_fr)
        
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
    
    # Pause for 15 seconds before listening again
    time.sleep(15)
```
