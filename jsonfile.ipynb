```python
import json
from googletrans import Translator

# Load the JSON file
with open('intents.json', 'r', encoding='utf-8') as json_file:
    json_data = json.load(json_file)

# Initialize the translator
translator = Translator()

# Function to recursively translate values
def translate(obj):
    if isinstance(obj, dict):
        return {translator.translate(key, src='en', dest='fr').text: translate(value) for key, value in obj.items()}
    elif isinstance(obj, list):
        return [translate(item) for item in obj]
    elif isinstance(obj, str):
        return translator.translate(obj, src='en', dest='fr').text
    else:
        return obj

# Translate the JSON data
translated_data = translate(json_data)

# Save the translated JSON data to a new file
with open('translated_file.json', 'w', encoding='utf-8') as output_file:
    json.dump(translated_data, output_file, ensure_ascii=False, indent=4)

print("Translation completed and saved to translated_file.json")
```
