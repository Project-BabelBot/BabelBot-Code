```python
import requests
import pandas as pd

params = {
    'access_key': '443c2fa2930637637c6579d0d4ac370b',  # Replace with your AviationStack API access key
    'departure_iata': 'YYC',  # Replace with Calgary airport code (IATA code)
    'airline_name': 'Air Canada',  # Replace with the airline name ('Air Canada' or 'WestJet')
    # You can also use 'arrival_iata' if you want flights arriving at Calgary
}

api_result = requests.get('http://api.aviationstack.com/v1/flights', params=params)
api_response = api_result.json()

# Check if 'data' exists in the response before creating the DataFrame
if 'data' in api_response:
    flight_data = api_response['data']
    # Create a DataFrame using pandas
    df = pd.DataFrame(flight_data)
    
    # Export data to Excel file named 'aircanada_flights.xlsx' or 'westjet_flights.xlsx'
    airline_name = params['airline_name'].lower().replace(' ', '_')
    file_name = f'{airline_name}_flights.xlsx'
    df.to_excel(file_name, index=False)
    print(f"Data successfully exported to '{file_name}'")
else:
    print("No 'data' found in the API response.")
```

```python
import requests

params = {
    'access_key': '443c2fa2930637637c6579d0d4ac370b'
}

api_result = requests.get('http://api.aviationstack.com/v1/flights', params=params)

if api_result.status_code == 200:  # Check if the request was successful
    api_response = api_result.json()
    
    # Print the entire API response data
    print(api_response)

    if 'data' in api_response:  # Check if 'data' key exists in the response
        for flight in api_response['data']:
            if flight.get('live') and not flight['live']['is_ground']:
                print(u'%s flight %s from %s (%s) to %s (%s) is in the air.' % (
                    flight['airline']['name'],
                    flight['flight']['iata'],
                    flight['departure']['airport'],
                    flight['departure']['iata'],
                    flight['arrival']['airport'],
                    flight['arrival']['iata']))
    else:
        print("No 'data' found in the API response.")
else:
    print("Failed to retrieve data. Status code:", api_result.status_code)
```
