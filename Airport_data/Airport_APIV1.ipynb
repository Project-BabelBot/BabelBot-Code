```python
import pandas as pd
import random
import string

def generate_flight_data(airlines, city_airport, num_flights):
    flight_data = []
    used_flight_numbers = set()

    for _ in range(num_flights):
        airline_name, airline_code = random.choice(list(airlines.items()))
        city = random.choice(list(city_airport.keys()))
        departure_time = f"{random.randint(0, 23):02d}:{random.randint(0, 59):02d}"  # Random departure time HH:MM
        gate = random.choice(string.ascii_uppercase)  # Random gate
        flight_number = f"FL{random.randint(1000, 9999)}"  # Flight number starting with FL and 4 random digits

        # Ensure flight number is unique
        while flight_number in used_flight_numbers:
            flight_number = f"FL{random.randint(1000, 9999)}"

        # Add the flight number to the set of used flight numbers
        used_flight_numbers.add(flight_number)

        # Get airport name corresponding to the city
        airport_name = ', '.join(city_airport.get(city, []))

        # Append the flight details to the list
        flight_data.append([flight_number, airline_name, airline_code, city, airport_name, departure_time, gate])

    # Create a Pandas DataFrame from the flight data
    flight_df = pd.DataFrame(flight_data, columns=["Flight Number", "Airline Name", "Airline Code", "City", "Airport Name", "Departure Time", "Gate"])

    return flight_df

# Example usage
airlines = {
    "Air Canada": "AC",
    "WestJet": "WS",
    "Flair": "Y9",
    # Add other airlines as needed
}

city_airport = {
    'Toronto': ['YYZ', 'YTZ'],
    'Vancouver': ['YVR'],
    'Edmonton': ['YEG'],
    # Add other city-airport mappings
}

num_flights_to_generate = 270
generated_flight_data = generate_flight_data(airlines, city_airport, num_flights_to_generate)

# Print generated flight data
print(generated_flight_data)

# Save flight data to Excel
generated_flight_data.to_excel("flight_details_with_airport.xlsx", index=False)
```
